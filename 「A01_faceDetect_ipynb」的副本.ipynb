```python
import cv2
from google.colab.patches import cv2_imshow
###   人臉偵測   ###
img = cv2.imread('pic001.jpg', cv2.IMREAD_UNCHANGED)
faceDetector=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
picFaces=faceDetector.detectMultiScale(img,scaleFactor=1.5,minNeighbors=4,minSize=(70,70),maxSize=(500,500))
print("在圖中辨識出人臉的個數為:",len(picFaces),"位")
###  顯示辨識後的圖片(加人臉框):
##將所有發現的人臉用方框標註:
for (x,y,w,h) in picFaces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
##在影像左上角標註偵測發現的人數:
cv2.rectangle(img,(5,5),(150,30),(0,255,255),-1)
cv2.putText(img,"Finding "+str(len(picFaces))+ " face(s)",
            (10,20),cv2.FONT_ITALIC,0.5,(255,0,0),2)
## 使用OpenCV視窗顯示影像: cv2.imshow(視窗名稱,影像物件)
cv2_imshow(img)
```
